```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1['date'] = df1['date'].str[:7]
```

```python
df1 = df1['date'].value_counts(sort= False).drop('2022-09',axis = 0).reset_index()
df1.head()
#.reset_index(name = 'counts')
```

```python
df1.columns = ['ds','y']
```

```python
from prophet import Prophet
```

```python
model = Prophet()
```

```python
model.fit(df1)
```

```python
future = model.make_future_dataframe(periods = 24,freq = 'MS')
forecast = model.predict(future)
```

```python
forecast[['ds','yhat']].tail()
```

```python
plt.figure(figsize = (12,8))
sns.lineplot(data = forecast,x = 'ds',y ='yhat')
plt.xlim('2022-09-01','2024-08-01')
plt.grid()
```
